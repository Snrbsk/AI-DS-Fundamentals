```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
ageList = [10,20,30,30,30,40,50,60,70,75]
weightList = [30,40,80,85,86,87,88,89,90,70]
numpyAge = np.array(ageList)
numpyWeight = np.array(weightList)
```

```python
#creating graph
plt.plot(numpyAge,numpyWeight,"r")
#writing labels
plt.xlabel("age")
plt.ylabel("weight")
plt.title("weight referrences with age")
```

```python
numpySeries1 = np.linspace(0,10,21)
numpySeries2 = numpySeries1**2
numpySeries3 = numpySeries1**3
plt.plot (numpySeries3)
```

```python
plt.plot(numpySeries1,numpySeries3)
```

```python
plt.plot(numpySeries1,numpySeries3,"--")
```

```python
#draw diff way
plt.plot(numpySeries1,numpySeries3,"*-")
```

```python
#draw diff graph
plt.subplot(1,2,1)
plt.plot(numpySeries1,numpySeries2,"b*-")
plt.subplot(1,2,2)
plt.plot(numpySeries1,numpySeries3,"r*-")
```

```python
#Create free canvas
myFigure = plt.figure()
#([])
figureAxes = myFigure.add_axes([0.4,0.4,0.8,0.8])
figureAxes.plot(numpySeries1,numpySeries2,"g")
figureAxes.set_xlabel("Number")
figureAxes.set_ylabel("Number^2")
```

```python
figure2 = plt.figure()
#([x start location,y start location, weight, height ])
eksen1 = figure2.add_axes([0.0,0.0,0.9,0.9])
eksen2 = figure2.add_axes([0.1,0.4,0.4,0.4])

eksen1.plot(numpySeries1,numpySeries3)
eksen2.plot(numpySeries1,numpySeries2)


eksen1.set_xlabel("Num")
eksen1.set_ylabel("Num^3")
eksen1.set_title("Kp Grafii")

eksen2.set_xlabel("Num")
eksen2.set_ylabel("Num^2")
eksen2.set_title("Kare Grafii")
```

```python
#Shortly draw
(myFigure,myAxis) = plt.subplots()
myAxis.plot(numpySeries1,numpySeries2,"g")
```

```python
#(nrows= stun grafik says, ncols= satr grafik says)
(myFigure,myAxes) = plt.subplots(nrows=1,ncols=4)
for ax in myAxes:
    ax.plot(numpySeries1,numpySeries2,"g")
    ax.set_xlabel("num")
    ax.set_ylabel("num^2")
    ax.set_title("Square")

#Grnty okunur hale getirir(uzaklatrr vs)
plt.tight_layout()
```

```python
newFigure = plt.figure()
newAxis = newFigure.add_axes([0.1,0.1,0.7,0.7])
#need for label-> legend necessary
newAxis.plot(numpySeries1,numpySeries3, label= "numpy series**2")
newAxis.legend()
```

```python
#Draw in same graph
(myFigure,myAxis) = plt.subplots()
myAxis.plot(numpySeries1,numpySeries2,"y--", label="numpy series^2")
myAxis.plot(numpySeries1,numpySeries3,"r*-", label="numpy series^3")
myAxis.legend(loc= 2)
```

```python
#myFigure = plt.figure(dpi=500) -> znrl arttrr
(myFigure,myAxis) = plt.subplots()
myAxis.plot(numpySeries1,numpySeries2,"y--", label="numpy series^2",linewidth = 4.0) #linewidth: izgi kalnl
myAxis.plot(numpySeries1,numpySeries3,"r*-", label="numpy series^3",alpha= 0.5) #alpha: transparency
myAxis.legend(loc= 2)
myFigure.savefig("myFigure.png",dpi=300)
```
